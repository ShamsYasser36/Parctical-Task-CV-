# What is YOLO?

 • YOLO (You Only Look Once) is a real-time 
object detection algorithm.

 • Identifies and locates multiple objects in an 
image or video in a single step.

 • It predicts bounding boxes and class labels 
simultaneously.

 • Key benefit: Faster and more efficient than 
previous detection algorithms.

# How YOLO Works ?

• 1. Divide Image: The image is divided into a grid.
    
 • 2. Prediction: Each grid cell predicts bounding 
boxes, class probabilities, and confidence scores.

 • 3. Bounding Boxes: Predicts the location and size 
of each object.
    
 • 4. Non-Maximum Suppression: Removes 
duplicate or overlapping bounding boxes

# YOLO _ DETECTION FOR PHOTO 

```python
 from ultralytics import YOLO
```

```python
 model = YOLO("yolov8n.pt")
```

```python
 results=model(source="photo_5963277072627517016_y.jpg",show=True
               ,conf=0.4,save=True)
```
