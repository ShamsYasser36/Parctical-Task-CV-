# What Image Segmentation? 

 Objective: The goal of image segmentation is to 
partition an image into regions that are more 
meaningful or easier to analyze, such as 
identifying specific objects, boundaries, or 
textures. 

```python
from ultralytics import YOLO
```

```python
import matplotlib.pyplot as plt 
```

```python
model = YOLO("yolov8n-seg.pt") 
```

```python
results = model("photo_5963277072627517016_y.jpg") 
```

```python
result_img = results[0].plot() 
plt.imshow(result_img) 
plt.axis('off') 
plt.show() 
```
